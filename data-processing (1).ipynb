```python
from scipy import stats as st
from sklearn.utils import shuffle
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('E:\\training.csv')
```

```python
data['Hypertension_medicine_years'] = data['Hypertension_medicine_years'].replace('VALID', np.NaN)
data['Hypertension_medicine_years'] = data.Hypertension_medicine_years.astype(float)
array1 = data['Hypertension_medicine_years'].replace(np.NaN,-1)
length = array1.shape[0]
# print(length)
arr1 = [];
for i in range(length):
    if array1[i] != -1:
        arr1.append(array1[i])  

# print(arr3)
trimmed =st.trim_mean(arr1, .05)
data['Hypertension_medicine_years'] = data['Hypertension_medicine_years'].fillna(data.apply(lambda cols: trimmed if str(cols.Hypertension).lower() == 'yes' and (str(cols.Hypertension_medicine_years).lower() == 'nan' or str(cols.Hypertension_medicine_years).lower() == 'no') else cols.Hypertension_medicine_years, axis=1))

data.head(20)
```

```python

data['Diabetes_medicine_years'] = data['Diabetes_medicine_years'].replace('VALID', np.NaN)
data['Diabetes_medicine_years'] = data.Diabetes_medicine_years.astype(float)
array2 = data['Diabetes_medicine_years'].replace(np.NaN,-1)
length = array2.shape[0]
# print(length)
arr2 = [];
for i in range(length):
    if array2[i] != -1:
        arr2.append(array2[i])  

# print(arr3)
trimmed = st.trim_mean(arr2, .05)
data['Diabetes_medicine_years'] = data['Diabetes_medicine_years'].fillna(data.apply(lambda cols: trimmed if str(cols.Diabetes).lower() == 'yes' and (str(cols.Diabetes_medicine_years).lower() == 'nan' or str(cols.Diabetes_medicine_years).lower() == 'no') else cols.Diabetes_medicine_years, axis=1))

data.head(20)
```

```python

data['Cholesterol_medicine_years'] = data['Cholesterol_medicine_years'].replace('VALID', np.NaN)
data['Cholesterol_medicine_years'] = data.Cholesterol_medicine_years.astype(float)
array3 = data['Cholesterol_medicine_years'].replace(np.NaN,-1)
length = array3.shape[0]
# print(length)
arr3 = [];
for i in range(length):
    if array3[i] != -1:
        arr3.append(array3[i])  

# print(arr3)
trimmed = st.trim_mean(arr3, .05)
data['Cholesterol_medicine_years'] = data['Cholesterol_medicine_years'].fillna(data.apply(lambda cols: trimmed if str(cols.Cholesterol).lower() == 'yes' and (str(cols.Cholesterol_medicine_years).lower() == 'nan' or str(cols.Cholesterol_medicine_years).lower() == 'no') else cols.Cholesterol_medicine_years, axis=1))

data.head(20)
```

```python

data['Heart_attack'] = data['Heart_attack'].replace(['Yes','No'],[1,0])
data['Maybe_heart_attack'] = data['Maybe_heart_attack'].replace(['Yes','No'],[1,0])
data['No_heart_attack'] = data['No_heart_attack'].replace(['Yes','No'],[1,0])
data=data.replace({'Yes':2,'No':3})
data = data.replace(np.NaN,3 ) 
#data.to_csv(r'E:\\data.csv')
data.head(10)
```

```python
data['Smoking'] =data['Smoking'].replace(['Ex-smoker'],4)
data['Gender'] =data['Gender'].replace(['Male','Female'],[4,5])
data['Chestpain_time'] =data['Chestpain_time'].replace(['10 am - 4 pm','4 pm - 10 pm','10 pm - 4 am','4 am - 10 am'],[4,5,6,7])
data ['Chestpain_location'] =data['Chestpain_location'].replace(['Center','Left side','Right side','Epigastric','Can not explain'],[4,5,6,7,8])
data ['Chestpain_mark'] =data['Chestpain_mark'].replace(['Fingure','Palm'],[4,5])
data['Doing_while_started']=data['Doing_while_started'].replace(['Working','Resting','Not remember'],[4,5,6])
data['Persistence']=data['Persistence'].replace(['<1min','1-5 min','>5 min but <20 min','>30 min but <1 hr','>1 hr','Whole day'],[4,5,6,7,8,9])
data ['Subsided']=data['Subsided'].replace(['By resting','Tablet/spray under tongue','After drinking water','Injection','Did not subsided'],[4,5,6,7,8])
data['Pain_type']=data['Pain_type'].replace(['Heaviness','Burning','Constricting','Could not breadth','Can not explain'],[4,5,6,7,8])
data['After_drinking_water_pain']=data['After_drinking_water_pain'].replace(['Increase','Decrease','No change'],[4,5,6])
data['After_having_food_pain']=data['After_having_food_pain'].replace(['Increase','Decrease','No change'],[4,5,6])
data ['By_chest_movement_pain'] =data['By_chest_movement_pain'].replace(['Increase','Decrease','No change'],[4,5,6])
data.to_csv(r'E:\\dataset1.csv')
data.head(10)
```
